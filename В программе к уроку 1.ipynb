## 4. Задание (в программе к уроку 1):
Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.


```python
x = np.linspace(-5,5)
y1 = np.cos(x)
y2 = np.cos(1.5*x)
plt.plot(x, y1, color='green', label='y1=cos(x)')
plt.plot(x, y2, label='y2=cos(1.5*x)')
plt.xlabel('Переменная x')
plt.ylabel('Переменная y')
plt.grid(color='lightgrey')
plt.legend(loc='lower right')
```

## 1. Задание
Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10)
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)


```python
x = np.array([10, 10, 10])
y = np.array([0, 0, -10])
c = np.add(x, y)
```

```python
def leng_arr(a):
    lengt=0
    for i in a:
        lengt += i**2
    lengt = math.sqrt(lengt)
    return lengt
```

```python
leng_arr(x)
```

```python
leng_arr(y)
```

```python
leng_arr(c)
```

## 5. Задание (в программе)
Нарисуйте трехмерный график двух параллельных плоскостей.
Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
X1 = np.arange(-5,5,0.5)
Y1 = np.arange(-5,5,0.5)
X1,Y1 = np.meshgrid(X1,Y1)
Z1=5*X1+2*Y1
ax.plot_wireframe(X1,Y1,Z1)
Z2=5*(X1-20)+2*(Y1+20)
ax.plot_wireframe(X1,Y1,Z1)
ax.plot_wireframe(X1,Y1,Z2)
show()
```

```python
fig = figure()
ax = Axes3D(fig)
X1 = np.arange(-5,5,0.5)
Y1 = np.arange(-5,5,0.5)
X1,Y1 = np.meshgrid(X1,Y1)
Z3=X1**2+Y1**2
Z4=(X1-2)**2-Y1**2
ax.plot_surface(X1,Y1,Z3)
ax.plot_surface(X1,Y1,Z4)
show()
```
